**Outlier Detection**

```python
Q1 = df['Rating'].quantile(0.25)
Q3 = df['Rating'].quantile(0.75)
IQR = Q3 - Q1
outliers = (df['Rating'] < Q1 - 1.5 * IQR) | (df['Rating'] > Q3 + 1.5 * IQR)
df = df[~outliers]
df
```

```python
import pandas as pd

# Check for non-year values in the 'Founded' column
non_year_values = df['Founded'].apply(lambda x: not pd.to_datetime(str(x), format="%Y", errors="coerce"))

# Print the number of non-year values
print(f"Number of non-year values: {non_year_values.sum()}")
```

```python
# Remove rows with non-year values
df = df[~non_year_values]
```

```python
# Convert all non-year values to 1900
df['Founded'] = df['Founded'].apply(lambda x: 1900 if not pd.to_datetime(str(x), format="%Y", errors="coerce") else x)
df
```

```python
# Convert 'Founded' column to datetime with errors='coerce'
df['Founded'] = pd.to_datetime(df['Founded'], format="%Y", errors='coerce')

# Remove rows with NaT values
df.dropna(subset=['Founded'], inplace=True)
```

```python
df
```

```python
# Convert 'Founded' column to datetime (example)
df['Founded'] = pd.to_datetime(df['Founded'], format='%Y')
```

**feature engineering **

```python
# Extract year from 'Founded' column
df['Founded_Year'] = df['Founded'].dt.year

# Calculate length of job descriptions
df['Description_Length'] = df['Job Description'].apply(len)
```

```python
# One-hot encoding with error handling
try:
    df = pd.get_dummies(df, columns=['Location', 'Company Name'])
except KeyError as e:
    print(f"Error: {e}")
df
```
